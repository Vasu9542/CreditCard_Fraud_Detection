```python
#Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#load the dataset to a pandas dataframe
cc_data=pd.read_csv("/content/creditcard.csv")
```

```python
cc_data.head()
```

```python
cc_data.tail()
```

```python
#Dataset information
cc_data.info()
```

```python
#Checking no. of missing values present in each column
cc_data.isnull().sum()
```

```python
creditcard_data=cc_data.dropna()
print(creditcard_data)
```

```python
creditcard_data.head()
```

```python
creditcard_data.isnull().sum()
```

```python
#Distribution of legit transaction and fraudulent transaction
creditcard_data['Class'].value_counts()
```

Dataset is highly unbalanced

0---> Normal Transaction

1---> Fraudulent Transaction


```python
#Separating data for analysis
legit=creditcard_data[creditcard_data.Class==0]
fraud=creditcard_data[creditcard_data.Class==1]
```

```python
legit.shape,fraud.shape
```

```python
#Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values for both transactions
creditcard_data.groupby('Class').mean()
```

```python
#Undersampling
#Build a sample dataset containing similar distribution of normal transaction and fraudulent transaction
#No. of Fraudulent transcations ---> 493
legit_sample=legit.sample(n=493)
```

```python
#Concatnating two dataFrames
N_Dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
N_Dataset.head()
```

```python
N_Dataset.tail()
```

```python
N_Dataset.info()
```

```python
N_Dataset['Class'].value_counts()
```

```python
N_Dataset.groupby('Class').mean()
```

```python
#Splitting the data into features and target
X=N_Dataset.drop(columns="Class",axis=1)
y=N_Dataset['Class']
```

```python
X, y
```

```python
#Split the data into training and testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python
#Model Training
#For binary classification we are using logistic regression
model=LogisticRegression()
```

```python
#Train the model
model.fit(X_train,y_train)
```

```python
#Evaluatuion of the model
#Accuracy score
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
```

```python
print("Accuracy on training data: ",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,y_test)
```

```python
print("Accuracy on test data: ",testing_data_accuracy)
```
